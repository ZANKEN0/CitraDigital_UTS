```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Muat gambar dalam mode grayscale
gambar = "./img/polkadot_besar_kecil.jpg"
gambar2 = "./img/buyung.jpg"

img = cv2.imread(gambar, 0)
img2 = cv2.imread(gambar2, 0)

# Ubah gambar ke binary (hitam putih)
_, img_binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
_, img_binary2 = cv2.threshold(img2, 127, 255, cv2.THRESH_BINARY)
# Buat Kernel berukuran 5x5
kernel = np.ones((3,3), np.uint8)
print("Kernel:\n", kernel)  

# Tampilkan gambar asli untuk perbandingan
plt.imshow(img_binary, cmap='gray')
plt.title('Gambar Asli')
plt.axis('off')
plt.show()
plt.imshow(img_binary2, cmap='gray')
plt.title('Gambar Asli')
plt.axis('off')
plt.show()
```

<h2>Macam-macam Karnel</h2>
<ul>
<li>KOTAK</li>
<li>BINTANG</li>
<li>TANDA PLUS</li>
</ul>

```python
rect_kernel = cv2 . getStructuringElement ( cv2 . MORPH_RECT , (5 ,5) )
ellipse_kernel = cv2 . getStructuringElement ( cv2 .MORPH_ELLIPSE , (5 ,5) )
cross_kernel = cv2 . getStructuringElement ( cv2 .MORPH_CROSS , (5 ,5) )

print ("Rectangular Kernel :\n" , rect_kernel )
print ("Elliptical Kernel :\n" , ellipse_kernel )
print ("Cross Kernel :\n", cross_kernel )
```

<h1>Saya Menggunakan karnel NUMPY dengan Ukuran 3x3</h1>

<h1>Erosion</h1>

<h2>Erosion: Mengecilkan objek, menghilangkan noise kecil
</h2>
<p>Pixel bernilai 1 akan
tetap 1 hanya jika semua pixel di bawah SE adalah 1.</p>

```python
# Terapkan Erosi
gambar_erosi = cv2.erode(img_binary, kernel, iterations=1)
gambar_erosi2 = cv2.erode(img_binary2, kernel, iterations=1)
# gambar_erosi1 = cv2.erode(img_binary, ellipse_kernel, iterations=1)
# gambar_erosi2 = cv2.erode(img_binary, cross_kernel, iterations=1)

plt.imshow(gambar_erosi, cmap='gray')
plt.title('Hasil Erosi: Objek Jadi Lebih Kurus')
plt.axis('off')
plt.show()


plt.imshow(gambar_erosi2, cmap='gray')
plt.title('Hasil Erosi: Objek Jadi Lebih Kurus')
plt.axis('off')
plt.show()
```

<h1>Dilation</h1>
<h2>Dilation: Memperbesar objek, menutup lubang</h2>
<p>Pixel bernilai 0 akan menjadi 1 jika setidaknya satu pixel di bawah SE adalah 1.
</p>

```python
# Terapkan Dilasi
gambar_dilasi = cv2.dilate(img_binary, kernel, iterations=1)
gambar_dilasi2 = cv2.dilate(img_binary2, kernel, iterations=1)

plt.imshow(gambar_dilasi, cmap='gray')
plt.title('Hasil Dilasi: Objek Jadi Lebih Gemuk')
plt.axis('off')
plt.show()
plt.imshow(gambar_dilasi2, cmap='gray')
plt.title('Hasil Dilasi: Objek Jadi Lebih Gemuk')
plt.axis('off')
plt.show()
```

<h1>Opening</h1>
<h2>Opening: Erosion + Dilation (noise removal di background)</h2>

```python
# Terapkan Opening
gambar_opening = cv2.morphologyEx(img_binary, cv2.MORPH_OPEN, kernel)
gambar_opening2 = cv2.morphologyEx(img_binary2, cv2.MORPH_OPEN, kernel)

plt.imshow(gambar_opening, cmap='gray')
plt.title('Hasil Opening: Noda Hilang, Objek Aman')
plt.axis('off')
plt.show()
plt.imshow(gambar_opening2, cmap='gray')
plt.title('Hasil Opening: Noda Hilang, Objek Aman')
plt.axis('off')
plt.show()

```

<h1>Closing</h1>
<h2>Closing: Dilation + Erosion (noise removal di foreground)</h2>


```python
# Terapkan Closing
gambar_closing = cv2.morphologyEx(img_binary, cv2.MORPH_CLOSE, kernel)
gambar_closing2 = cv2.morphologyEx(img_binary2, cv2.MORPH_CLOSE, kernel)

plt.imshow(gambar_closing, cmap='gray')
plt.title('Hasil Closing: Lubang di Dalam Objek Tertutup')
plt.axis('off')
plt.show()

plt.imshow(gambar_closing2, cmap='gray')
plt.title('Hasil Closing: Lubang di Dalam Objek Tertutup')
plt.axis('off')
plt.show()
```

<h3> MORPHOLOGICAL GRADIENT </h3>

```python
gradient = cv2.morphologyEx(img_binary,cv2.MORPH_GRADIENT,kernel)
gradient2 = cv2.morphologyEx(img_binary2,cv2.MORPH_GRADIENT,kernel)

plt.imshow(gradient, cmap='gray')
plt.title('Hasil Gradient')
plt.axis('off')
plt.show()
plt.imshow(gradient2, cmap='gray')
plt.title('Hasil Gradient')
plt.axis('off')
plt.show()
```
